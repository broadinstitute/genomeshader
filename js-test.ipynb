```python
from IPython.display import display, HTML

# HTML and JavaScript code to open a new window, draw a red rectangle, and display GPU information below it
html_script = """
<script type="text/javascript">
(function() {
    var newWindow = window.open("", "newWindow", "width=600,height=650");
    if (!newWindow) return;
    
    // Adjust the layout to ensure the GPU information is displayed below the canvas
    var contentHtml = '<canvas id="myCanvas" width="400" height="400" style="border:1px solid #000; display: block; margin-bottom: 20px;"></canvas><div id="gpuInfo" style="font-family: Arial, sans-serif;"></div>';
    newWindow.document.body.innerHTML = contentHtml;
    
    var canvas = newWindow.document.getElementById('myCanvas');
    var context = canvas.getContext('2d');
    
    // Draw a red rectangle
    context.fillStyle = "#FF0000"; // Set fill color to red
    context.fillRect(50, 50, 300, 300); // Fill a rectangle

    // GPU information extraction
    var gl;
    var debugInfo;
    var vendor;
    var renderer;

    try {
      gl = canvas.getContext('webgl') || canvas.getContext('experimental-webgl');
    } catch (e) {
    }

    if (gl) {
      debugInfo = gl.getExtension('WEBGL_debug_renderer_info');
      vendor = gl.getParameter(debugInfo.UNMASKED_VENDOR_WEBGL);
      renderer = gl.getParameter(debugInfo.UNMASKED_RENDERER_WEBGL);
      
      // Display GPU Information clearly below the canvas
      var gpuInfoDiv = newWindow.document.getElementById('gpuInfo');
      gpuInfoDiv.innerHTML = '<p><strong>GPU Vendor:</strong> ' + vendor + '</p><p><strong>GPU Renderer:</strong> ' + renderer + '</p>';
    }
})();
</script>
"""

# Display the HTML and JavaScript
display(HTML(html_script))
```

```python
from IPython.display import display, HTML

# HTML and JavaScript code to open a new window, display GPU information on the left, and use WebGPU on the right
html_script = """
<script type="text/javascript">
(async function() {
    var newWindow = window.open("", "newWindow", "width=800,height=400");
    if (!newWindow) return;
    
    // Layout for GPU info and WebGPU rendering
    var contentHtml = '<div style="display: flex; height: 100%; font-family: Arial, sans-serif;">' +
                      '<div id="gpuInfo" style="flex: 1; display: flex; align-items: center; justify-content: center; text-align: center; padding: 20px;"></div>' +
                      '<canvas id="webgpuCanvas" style="flex: 1; width: 50%; height: 100%;"></canvas>' +
                      '</div>';
    newWindow.document.body.innerHTML = contentHtml;

    var canvas = newWindow.document.getElementById('webgpuCanvas');

    if (!('gpu' in navigator)) {
      newWindow.document.getElementById('gpuInfo').innerHTML = '<p>WebGPU is not supported by your browser.</p>';
      return;
    }
    
    const adapter = await navigator.gpu.requestAdapter();
    const device = await adapter.requestDevice();
    
    // Display generic GPU Information (detailed info like vendor/renderer is not directly available in WebGPU)
    var gpuInfoDiv = newWindow.document.getElementById('gpuInfo');
    gpuInfoDiv.innerHTML = '<p>WebGPU is supported. Rendering with WebGPU.</p>';
    
    // Setup WebGPU for the canvas
    const context = canvas.getContext('webgpu');
    const format = 'bgra8unorm';
    context.configure({
        device: device,
        format: format,
        alphaMode: 'opaque',
    });

    // Create a view for the entire size of the canvas
    const view = context.getCurrentTexture().createView();

    // Commands to clear the canvas to dark blue
    const commandEncoder = device.createCommandEncoder();
    const renderPassDescriptor = {
        colorAttachments: [{
            view: view,
            clearValue: {r: 0.3, g: 0, b: 0.5, a: 1}, // Dark blue
            loadOp: 'clear',
            storeOp: 'store',
        }],
    };
    const passEncoder = commandEncoder.beginRenderPass(renderPassDescriptor);
    //passEncoder.endPass();

    // Submit the commands to the GPU
    device.queue.submit([commandEncoder.finish()]);
})();
</script>
"""

# Display the HTML and JavaScript
display(HTML(html_script))
```
